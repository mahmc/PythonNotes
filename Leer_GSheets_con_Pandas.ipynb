## Introducción

Mostramos una manera de conectar un Google Sheets con Google Colab y convertirlo en un dataframe.

## Módulos necesarios

Utilizamos los módulos `google` y `gspread` para conectar un Google Sheets con Google Colab.

```python
from google.colab import auth
from google.auth import default
import gspread
import pandas as pd
```

Autentica al usuario utilizado en Google Colab, para acceder a los Google Sheets que tiene permiso.


Al ejecutar la siguiente celda de código se muestra un mensaje (ver imagen), donde tiene que dar permisos para que Google Colab pueda acceder a los archivos de Google Sheets.

Screenshot 2023-10-10 at 21.59.20.png

```python
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

Con los permisos necesarios, ya podemos acceder a un documento aolojado en Google Sheets, para esto es necesario solo agregar la url del Google Sheets y la sheet a utilizar.   

```python
gsheets_url = "https://docs.google.com/spreadsheets/d/1I7eHy0V93FkIMnUSsL2nUeX3tr9SvM8yDBDK6fg4JDg/edit#gid=0"
sheet = 'test'
```

Con los datos anteriores ya podemos utilizar el sheet y guardalos en un variable de Python.

```python
gsheets = gc.open_by_url(gsheets_url)
sheet = gsheets.worksheet(sheet).get_all_values()
```

La variable `sheet` es una lista con dos elementos. El primer elementos son el nombre las columnas y el segundo elemento en una lista de lista que contiene los datos asociados a las columnas.
